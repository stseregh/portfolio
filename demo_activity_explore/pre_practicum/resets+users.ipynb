```python
import pandas as pd
resets = pd.read_excel("main template resets.xlsx")
users = pd.read_excel("Downloads/Users Wrike.xlsx")
owners = pd.read_csv("Downloads/BQ data/demo acc owners.csv")
#matching demo onwer IDs to Acc IDs
resets = pd.merge(resets,owners, how="left", left_on='Account',right_on='account_id')
#add user metadata
resets = pd.merge (resets,users,how='left',left_on='Owner',right_on='ID')
#clean up a bit
resets.drop('ID', axis=1, inplace=True)
resets.drop('account_id', axis=1, inplace=True)
resets
```

```python
resets.to_excel("main template resets full.xlsx", sheet_name="resets", index=False)
```
